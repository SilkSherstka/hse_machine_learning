## ЗАДАНИЕ 1: Базовые графики matplotlib

```python
# Создайте фигуру с 4 subplots (2x2):
# 1. Линейный график функции y = x^2 * sin(x) на интервале [0, 10]
#    Добавьте заголовок, подписи осей, сетку, легенду
# 2. Гистограмма 1000 случайных точек из нормального распределения
#    Настройте количество bins=30, добавьте заголовок
# 3. Scatter plot: 50 точек (x - равномерное распределение, y = 2x + шум)
#    Настройте размер точек, прозрачность, цвет
# 4. Bar chart: сравнение 5 разных моделей по точности
#    accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
#    Добавьте значения над столбцами

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка стиля seaborn для более приятных графиков по умолчанию
sns.set_style("whitegrid")

# Настройка отображения графиков внутри ноутбука
%matplotlib inline

# Настройка размеров шрифтов для лучшей читаемости
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10
})

# Создаем фигуру с 4 subplots (2x2)
### ваш код здесь ###
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
########################################
# 1. Линейный график y = x^2 * sin(x)
ax = axes[0, 0]
x = np.linspace(0, 10, 100)
y = x**2 * np.sin(x)
### ваш код здесь ###
ax.plot(x, y, 'b-', linewidth=2, label='y = x^2 * sin(x)')

# Добавление заголовка
ax.set_title('y = x^2 * sin(x)', fontsize=14)

# Подписи осей (обязательны для читаемости)
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Включение сетки
ax.grid(True, alpha=0.3)  # alpha - прозрачность

# Добавление легенды
ax.legend()
#############################################
# 2. Гистограмма нормального распределения
ax = axes[0, 1]
data = np.random.normal(0, 1, 1000)
### ваш код здесь ###
ax.hist(data, bins=30, edgecolor='black', alpha=0.7, color='lightblue')

# Добавление заголовка
ax.set_title('Гистограмма (1000 точек)', fontsize=14)

# Подписи осей (обязательны для читаемости)
ax.set_xlabel('Значение')
ax.set_ylabel('Частота')


# Включение сетки
ax.grid(True, alpha=0.3)  # alpha - прозрачность

#############################################

# 3. Scatter plot: y = 2x + шум
ax = axes[1, 0]
x = np.random.uniform(0, 10, 50)
y = 2 * x + np.random.normal(0, 1, 50)
### ваш код здесь ###

ax.scatter(x, y, s=50, alpha=0.7, color='green', edgecolors='green', linewidth=1)

# Добавление заголовка
ax.set_title('Scatter plot: y = 2x + шум', fontsize=14)

# Подписи осей (обязательны для читаемости)
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Включение сетки
ax.grid(True, alpha=0.3)  # alpha - прозрачность

#############################################

# 4. Bar chart: точность моделей
ax = axes[1, 1]
models = ['A', 'B', 'C', 'D', 'E']
accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
### ваш код здесь ###

bars = ax.bar(models, accuracy, color='skyblue', linewidth=1.5)

# Добавляем значения над столбцами
for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
            f'{acc:.2f}', ha='center')
# Добавление заголовка
ax.set_title('Сравнение моделей', fontsize=14)

# Подписи осей (обязательны для читаемости)
ax.set_xlabel('Модель')
ax.set_ylabel('Accuracy')

# Включение сетки
ax.grid(True, alpha=0.3)  # alpha - прозрачность

plt.tight_layout()
plt.show()
```

## ЗАДАНИЕ 2: Seaborn визуализации

```python
# Используйте датасет tips из seaborn:
# import seaborn as sns
# tips = sns.load_dataset('tips')

# Создайте 3 графика:
# 1. Box plot: total_bill по дням недели (day)
#    Разделите дополнительно по времени (time) с помощью hue
# 2. Violin plot: tip в зависимости от размера компании (size)
# 3. Pair plot: выберите 4 числовых признака, раскрасьте по полу (sex)

import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет
tips = sns.load_dataset('tips')

# Создаем фигуру для графиков
### ваш код здесь ###
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
ax1 = axes[0]


# 1. Box plot: total_bill по дням с разделением по времени
#data=tips, x='day', y='total_bill', hue='time'
### ваш код здесь ###
box = sns.boxplot(data=tips, x='day', y='total_bill', hue='time',
                  palette='Set2', ax=ax1)

ax1.set_title('Total bill по дням и времени', fontsize=14)
ax1.set_xlabel('День недели', fontsize=12)
ax1.set_ylabel('Total bill', fontsize=12)
ax1.legend(title='time', loc='upper center')
ax1.grid(True, alpha=0.3, axis='y')

medians = tips.groupby(['day', 'time'])['total_bill'].median().reset_index()
for i, row in medians.iterrows():
    day_order = ['Thur', 'Fri', 'Sat', 'Sun']
    day_idx = day_order.index(row['day'])

    ax1.text(day_idx, row['total_bill'] + 0.5,
             f'{row["total_bill"]:.1f}',
             ha='center', va='bottom', fontsize=9, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.7))


# 2. Violin plot: tip по размеру компании
# x='size', y='tip'
### ваш код здесь ###
sns.violinplot(data=tips, x='size', y='tip', inner='stick', ax=axes[1])


axes[1].set_title('Tip по размеру компании')
axes[1].set_xlabel('Размер компании')
axes[1].set_ylabel('Tip')


# 3. Pair plot: 4 числовых признака, раскраска по полу
# tips[['total_bill', 'tip', 'size', 'sex']]
# Создаем отдельную фигуру для pair plot
### ваш код здесь ###
plt.figure(figsize=(12, 10))

pair_grid = sns.pairplot(tips[['total_bill', 'tip', 'size', 'sex']],
                         hue='sex',  # Раскраска по полу
                         palette={'Male': 'blue', 'Female': 'red'},  # Цвета для категорий
                         diag_kind='kde',
                         plot_kws={'alpha': 0.7, 's': 50, 'edgecolor': 'black'},
                         diag_kws={'alpha': 0.7, 'fill': True},
                         height=2.5)
pair_grid.fig.suptitle('Pair Plot по полу', fontsize=16, y=1.02)
```
