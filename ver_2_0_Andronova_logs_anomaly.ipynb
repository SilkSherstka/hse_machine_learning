



Выбранная модель: Isolation Forest

Не требует размеченных данных - работает в режиме one-class (учится только на "нормальных" данных)
Работает с разными типами метрик - и временные ряды, и IP-адреса, и ошибки
Дает score аномальности - можно настроить порог чувствительности
Не требует нормализации данных - устойчив к масштабу признаков

Структура модулей:

```
log-anomaly-detector/           
├── README.md                    
├── config.yaml                  #Конфиг для параметров модели, алертов, и тд
│
├── data/                        # Директория для хранения данных
│   ├── raw/                     # Сырые логи в исходном формате
│   │   ├── 2024-01-01-00-00.log # Логи
│   │   └── 2024-01-02-00-01.log # Логи
│   ├── processed/                # Обработанные данные, готовые для обучения
│   │   ├── train_features.npy    # Матрица признаков для обучения
│   └── external/                  # Внешние данные для обогащения
│       └── known_ips.txt          # Белый список IP-адресов
│
├── src/                           # Исходный код проекта
│   ├── data/                       # Модули для работы с данными
│   │   ├── loader.py                # Загрузка логов из файлов, Kafka, API
│   │   ├── parser.py                # Парсинг различных форматов логов (JSON, CSV, plain text)
│   │   ├── preprocessor.py          # Очистка, фильтрация, нормализация данных
│   │   └── feature_engineering.py   # Создание признаков для модели (RPS, CPU, ошибки и т.д.)
│   │
│   ├── models/                      # Модули для работы с моделями ML
│   │   ├── isolation_forest.py       # Реализация Isolation Forest
│   │   ├── train.py                  # Скрипты для обучения модели
│   │   ├── predict.py                 # Скрипты для инференса (предсказаний)
│   │   └── evaluator.py               # Оценка качества модели (метрики, кросс-валидация)
│   │
│   │
│   ├── alerts/                        # Система оповещений
│   │   ├── generator.py                # Генерация алертов при обнаружении аномалий
│   │   ├── formatter.py                 # Форматирование сообщений для разных каналов
│   │   └── channels.py                  # Каналы отправки (console, slack, email, telegram)
│   │
│   └── monitoring/                     # Мониторинг работы системы
│       ├── logger.py                    # Логирование всех проверок и событий
│       └── metrics.py                    # Сбор метрик работы модели (латентность, точность)
│
└── logs/                             # Логи работы приложения
    ├── app.log                        # Общий лог приложения
    ├── alerts.log                      # Лог всех сгенерированных алертов
    └── model_predictions.log            # Лог всех предсказаний модели```



Роли:

1) Тимлид - ML команда
2) Аналитик (тесты, презентация, коммуникация) - ML команда
3) ML Engineer - ML команда
4) Data Engineer - ML команда
5) Аналитик - support-команда

Таймлайн
1. 1-2 недели - запуск модели, но с тестовым алертингом, только на команду разработки детектора. Оцениваем результат, фиксим очевидные баги, тюним модель
2. Презентуем инструмент в профильные команды, которым будет приходить алерт. Договариваемся о совместном тестовом режиме: запускаем тестовый алертинг на целевые команды, получаем фидбек, тюним модель
3. Следующие 2-3 недели живем в тестовом режиме. Принимаем решение о запуске, запуск целевого алертинга
